## 散点图：plt.scatter(x,y)

```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt  #显示中文设置
plt.rcParams['font.family'] = ['Microsoft YaHei']
```

```python
x=[1,2,3,4,6,7,9,2]
y=[2,3,4,5,2,4,6,7]
#创建画布
plt.figure(figsize=(20,8),dpi=100)

#图像绘制
plt.scatter(x,y)

#图像显示
plt.show()
```

## 柱状图：plt.bar(x, width, aligh="center",**kwargs)

```python
movie_name={'A','B','C','D'}
x=range(len(movie_name))
y=[1,2,3,4]

#创建画布
plt.figure(figsize=(20,8),dpi=100)

#图像绘制
plt.bar(x,y,color=['b','r','g','y'],width=0.5)

#x轴
plt.xticks(x,movie_name,fontsize=20)
#网格
plt.grid()
#标题
plt.title("柱状图案例",fontsize=20)
#图像显示
plt.show()
```
