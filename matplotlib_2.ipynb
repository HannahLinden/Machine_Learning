# 案例显示温度变化


```python
# 生成数据
x=range(60)
y_brijing=[random.uniform(10,15) for i in x]
y_shanghai=[random.uniform(20,25) for i in x]
# 创建画布
plt.figure(figsize=(20,8),dpi=100)

#图形绘制,plot为折线图
plt.plot(x,y_brijing,label="北京",color="g",linestyle="-.")
plt.plot(x,y_shanghai, label="上海")

# 添加想x，y轴刻度
y_ticks=range(40)  
x_ticks_lables=["11时{}m".format(i) for i in x]
plt.yticks(y_ticks[::5])
plt.xticks(x[::5],x_ticks_lables[::5])

#添加网格显示
plt.grid(True, linestyle="-",alpha=1)

#添加描述
plt.xlabel("时间",fontsize=20)
plt.ylabel("温度",fontsize=20)
plt.title("一小时温度变化图", fontsize=20)

# 显示图例，用legend显示，其中要显示的要添加标签label，9/10两行代码
plt.legend()  
#图像显示
plt.show()
```

# 案例二：多个坐标系显示图像

```python
# 生成数据
x=range(60)
y_brijing=[random.uniform(10,15) for i in x]
y_shanghai=[random.uniform(20,25) for i in x]
# 创建画布
plt.figure(figsize=(20,8),dpi=100)
fig,axes=plt.subplots(nrows=1, ncols=2,figsize=(20,8),dpi=100 )

#图形绘制,plot为折线图
# plt.plot(x,y_brijing,label="北京",color="g",linestyle="-.")
# plt.plot(x,y_shanghai, label="上海")
axes[0].plot(x,y_brijing,label="北京",color="g",linestyle="-.")
axes[1].plot(x,y_shanghai, label="上海",color="r",linestyle="-")

# # 添加想x，y轴刻度
y_ticks=range(40)  
x_ticks_lables=["11时{}分".format(i) for i in x]
# plt.yticks(y_ticks[::5])
# plt.xticks(x[::5],x_ticks_lables[::5])

axes[0].set_xticks(x[::5])
axes[0].set_yticks(y_ticks[::5])
axes[0].set_xticklabels(x_ticks_lables[::5])
axes[1].set_xticks(x[::5])
axes[1].set_yticks(y_ticks[::5])
axes[1].set_xticklabels(x_ticks_lables[::5])

# #添加网格显示
# plt.grid(True, linestyle="-",alpha=1)

# #添加描述
# plt.xlabel("时间",fontsize=20)
# plt.ylabel("温度",fontsize=20)
# plt.title("一小时温度变化图", fontsize=20)

axes[0].set_xlabel("时间",fontsize=20)
axes[0].set_ylabel("温度",fontsize=20)
axes[0].set_title("一小时温度变化图", fontsize=20)

axes[1].set_xlabel("时间",fontsize=20)
axes[1].set_ylabel("温度",fontsize=20)
axes[1].set_title("一小时温度变化图", fontsize=20)
# 显示图例，用legend显示，其中要显示的要添加标签label，9/10两行代码
# plt.legend()  
axes[0].legend()  
axes[1].legend()  
#图像显示
plt.show()
```

# 案例三plot绘制数学图像：sin、x^3

```python
import numpy as np
```

```python
# 生成数据
x = np.linspace(-10,10,1000)
y=np.sin(x)
y=x*x*x
#创建画布
plt.figure(figsize=(20,8),dpi=100)

#绘制图像
plt.plot(x,y)

plt.grid(True,linestyle="-",alpha=1)
#显示画布
plt.show()
```
