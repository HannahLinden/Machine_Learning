# 生成均匀分布的案例

```python
#准备数据
x=np.random.uniform(0,1,100000000)

#创建画布
plt.figure(figsize=(20,8),dpi=100)

#绘制
plt.hist(x,bins=1000)

#显示

plt.show()
```

# 生成正态分布数据

```python
x=np.random.normal(1.75,1,100000000)

#创建画布
plt.figure(figsize=(20,8),dpi=100)

#绘制
plt.hist(x,bins=1000)

#显示
plt.show()
```
